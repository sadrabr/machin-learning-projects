````python
# TODO: متد زیر را برای نشان دادن بار پلات تکمیل و متغیر های ذکر شده را بررسی کنید.
def plot_categorical_bars(df,
                        categorical_columns,
                        rot=None,  # پارامتر چرخش برای برچسب‌های محور X
                        bar_width=0.5,  # عرض میله‌ها
                        palette='dark',  # پالت رنگ
                        figsize=(16, 6)):  # اندازه شکل
    # تنظیم استایل Seaborn برای نمودار
    sns.set(style="darkgrid")

    # ایجاد زیرنمودارها با تعداد ردیف 1 و تعداد ستون‌ها برابر با تعداد متغیرها
    num_columns = len(categorical_columns)
    fig, axes = plt.subplots(1, num_columns, figsize=figsize)

    # صاف کردن آرایه axes برای دسترسی راحت‌تر به هر subplot
    # و  حلقه برای ایجاد نمودار میله‌ای برای هر ستون
    axes = axes.flatten()
    for i, col in enumerate(categorical_columns):
        ax = axes[i]
        sns.countplot(x=col, data=df, width=bar_width,
                      palette=palette, ax=ax)  # ایجاد نمودار میله‌ای برای هر ستون دسته‌بندی شده
        ax.set_title(f'Distribution of {col}')
        ax.set_xlabel(col)
        ax.set_ylabel('Count')
        # چرخاندن لیبل ها محور X
        if rot:
            ax.set_xticklabels(ax.get_xticklabels(), rotation=rot, ha='right')
        # نوشتن تعداد هر میله بالای آن برای نمایش
        for p in ax.patches:
            ax.annotate(format(p.get_height(), '.0f'),
                        (p.get_x() + p.get_width() / 2., p.get_height()),
                        ha='center', va='center',
                        xytext=(0, 5),
                        textcoords='offset points',
                        fontweight='bold',
                        color='black')
    # تنظیم فضای نمودارها برای جلوگیری از تداخل
    plt.tight_layout()
    plt.show()
````

````python
# TODO: متد زیر را برای نشان دادن هیستوگرام تکمیل و متغیر های ذکر شده را بررسی کنید.
def hist_plot(df, numeric_columns, bins=30, rot=None, figsize=(14, 8)):
    """
    این تابع چندین متغیر عددی را به صورت هستوگرام نمایش می‌دهد.
    پارامترها:
    df : DataFrame -> دیتافریم ورودی
    numeric_columns : list -> لیست متغیرهای عددی
    bins : int -> تعداد سطل‌های (bins) نمودار
    rot : int -> مقدار چرخش برچسب‌های محور X
    figsize : tuple -> اندازه نمودار
    """
    # تنظیم استایل Seaborn
    sns.set_style("darkgrid")

    # تعداد متغیرهای عددی
    # مشخص کردن تعداد سطرها و ستون‌ها برای نمودارها
    num_vars = len(numeric_columns)
    rows = (num_vars // 3) + (num_vars % 3 > 0)  # تقسیم تعداد متغیرها بر 3 و رند کردن به بالا
    cols = min(3, num_vars)

    # ایجاد subplot با تعداد مناسب
    fig, axes = plt.subplots(rows, cols, figsize=figsize)
    axes = axes.flatten()  # تبدیل آرایه 2D به 1D برای دسترسی آسان

    # حلقه برای رسم هستوگرام برای هر متغیر عددی
    for i, col in enumerate(numeric_columns):
        ax = axes[i]  # گرفتن subplot متناظر
        sns.histplot(data=df, x=col, bins=bins, ax=ax, kde=True, palette="viridis")  # رسم هستوگرام
        # تنظیم عنوان و برچسب‌ها
        ax.set_title(f'Distribution {col}')
        ax.set_xlabel(col)
        ax.set_ylabel('Count')

        # تنظیم چرخش برچسب‌های محور X
        if rot:
            ax.set_xticklabels(ax.get_xticklabels(), rotation=rot)

    # حذف subplot های اضافی در صورتی که تعداد متغیرها کمتر از تعداد کل subplot ها باشد
    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])

    # تنظیمات نهایی برای جلوگیری از تداخل
    plt.tight_layout()
    plt.show()
````

````python
# متغیر های مشخص شده برای Kde Plot
numeric_cols = ['Monthly Charges', 'Total Charges', 'Tenure Months', 'CLTV',]
````

````python
# TODO: متد زیر را برای نشان دادن KDE Plot تکمیل و متغیر های ذکر شده را بررسی کنید.
def plot_numeric_distributions(df, numeric_columns, target_col='Churn Label'):
    """
    رسم نمودارهای KDE برای متغیرهای عددی با تفکیک بر اساس متغیر هدف.
    این تابع توزیع متغیرهای عددی را برای هر مقدار از متغیر هدف (مانند 'Churn Label')
    در یک شبکه ۲×۲ نمایش می‌دهد.
    پارامترها:
    df (DataFrame): دیتافریم شامل داده‌ها
    numeric_cols (list): لیست ستون‌های عددی برای رسم نمودار
    target_col (str): نام ستون هدف برای دسته‌بندی (پیش‌فرض 'Churn Label')
    """
    fig, axes = plt.subplots(2, 2, figsize=(12, 8))

    axes = axes.flatten()  # تبدیل آرایه ۲ بعدی به ۱ بعدی برای دسترسی آسان‌تر

    for i, col in enumerate(numeric_columns):
        sns.kdeplot(df[df[target_col] == 'Yes'][col], label='Churn', shade=True, color='red', ax=axes[i])
        sns.kdeplot(df[df[target_col] == 'No'][col], label='No Churn', shade=True, color='blue', ax=axes[i])
        axes[i].set_title(f'Distribution of {col} by {target_col}')
        axes[i].legend()

    # حذف ساب‌پلات‌های خالی در صورتی که تعداد متغیرها فرد باشد
    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout()
    plt.show()
````

````python
# TODO: متد زیر را تکمیل کنید برای بررسی ارتباط متغیرهای دسته‌ای با  با هدف.
def plot_categorical_subplots(df,
                              categorical_cols,
                              target_col='Churn Label'):
    """
    رسم نمودارهای Countplot برای متغیرهای دسته‌ای همراه با متغیر هدف.
    این تابع نمودارهای شمارشی را برای هر متغیر دسته‌ای مشخص‌شده در کنار متغیر هدف
    (مانند 'Churn Label') در یک شبکه ۲×۲ نمایش می‌دهد.
    پارامترها:
    df (DataFrame): دیتافریم شامل داده‌ها
    categorical_cols (list): لیست ستون‌های دسته‌ای برای رسم نمودار
    target_col (str): نام ستون هدف برای نمایش hue (پیش‌فرض 'Churn Label')
    """
    num_cols = len(categorical_cols)
    num_rows = (num_cols // 2) + (num_cols % 2)  # تعیین تعداد ردیف‌ها برای 2x2
    fig, axes = plt.subplots(num_rows, 2, figsize=(14, num_rows * 5))

    axes = axes.flatten()  # تبدیل آرایه ۲ بعدی به ۱ بعدی برای دسترسی آسان‌تر

    for i, col in enumerate(categorical_cols):
        sns.countplot(x=col, hue=target_col, data=df, palette='Set2', ax=axes[i])
        axes[i].set_title(f'Churn Rate by {col}')
        axes[i].tick_params(axis='x', rotation=45)

    # حذف ساب‌پلات‌های خالی در صورتی که تعداد متغیرها فرد باشد
    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout()
    plt.show()
````

````python
# TODO: آماده سازی داده های برای استفاده از ColumnTransformer

# # تبدیل مقادیر هدف (y) از 'No' و 'Yes' به 0 و 1
y_train = y_train.map({'No': 0, 'Yes': 1})
y_test = y_test.map({'No': 0, 'Yes': 1})

numeric_features = X.select_dtypes(include=['int64', 'float64']).columns
categorical_features = X.select_dtypes(include=['object']).columns

# ایحاد یک پراسسور برای ترانسفورم های مورد نیاز هر ویژگی
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features),
        ('cat', OneHotEncoder(), categorical_features)
    ])
````

````python
# TODO: ایجاد دیکشنری از مدل های مورد نظر برای آموزش و ارزیابی اولیه
# TODO: آموزش داده با استفاده از پایپلاین
# TODO: ارزیابی داده ها روی مجموعه آموزش و تست

# دیکشنری مدل ها
models = {
    'Logistic Regression': LogisticRegression(),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'Gradient Boosting': GradientBoostingClassifier(),
    'XGBoost': XGBClassifier(),
    'AdaBoost': AdaBoostClassifier()
}

# آموزش و ارزیابی مدل
for model_name, model in models.items():
    print(f"Training {model_name}...")

    # ساخت پایپلاین
    pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('classifier', model)
    ])

    # آموزش مدل
    pipeline.fit(X_train, y_train)

    # پیشبینی
    y_train_pred = pipeline.predict(X_train)
    y_test_pred = pipeline.predict(X_test)

    # ارزیابی مدل روی داده های آموزشی
    print(f"Performance on Train Data for {model_name}:")
    print(f"Accuracy: {accuracy_score(y_train, y_train_pred)}")
    print(f"Classification Report:\n{classification_report(y_train, y_train_pred)}")
    print(f"Confusion Matrix:\n{confusion_matrix(y_train, y_train_pred)}")

    # ارزیابی مدل روی داده های آزمایشی
    print(f"\nPerformance on Test Data for {model_name}:")
    print(f"Accuracy: {accuracy_score(y_test, y_test_pred)}")
    print(f"Classification Report:\n{classification_report(y_test, y_test_pred)}")
    print(f"Confusion Matrix:\n{confusion_matrix(y_test, y_test_pred)}")

    print("-" * 50)
````

````python
numeric_features = X.select_dtypes(include=['int64', 'float64']).columns
categorical_features = X.select_dtypes(include=['object']).columns

# ایجاد پیش‌پردازش برای ویژگی‌های عددی
numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())  # اسکیل کردن داده‌ها
])

# ایجاد پیش‌پردازش برای ویژگی‌های دسته‌ای
categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))  # انکود کردن داده‌های دسته‌ای
])

# ترکیب پیش‌پردازش‌ها برای ویژگی‌های عددی و دسته‌ای
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ]
)

# ایجاد پایپلاین برای هر مدل
def create_model_pipeline(model):
    return Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('model', model)
    ])
````

````python
# تعریف تابع برای نمایش classification report و confusion matrix
def evaluate_model(model, X_test, y_test):
    # پیش‌بینی نتایج
    y_pred = model.predict(X_test)

    print("Classification Report:")
    print(classification_report(y_test, y_pred))
    print("Confusion Matrix:")
    cm = confusion_matrix(y_test, y_pred)
    print(cm)

    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["Class 0", "Class 1"], yticklabels=["Class 0", "Class 1"])
    plt.title("Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.show()
````
